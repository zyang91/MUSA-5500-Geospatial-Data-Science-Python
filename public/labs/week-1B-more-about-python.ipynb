## Python Functions

A function in Python is a block of code that only runs when it is called. You can pass data, known as parameters, into a function. A function can return data as a result or not return anything just conduct some tasks.

---

### Python functions structure

```python
def function_name(arg1, arg2, arg3):
    
    .
    .
    .
    code lines (indented)
    .
    .
    .
    #this line is optional
    return result 
```

---

### Defining a Function

To define a function, use the `def` keyword followed by the function name and parentheses. Here’s an example:

````python
def compute_square(x):
    
    sq = x * x
    return sq

sq = compute_square(10)
print(sq)
````

---

### Keywords:  arguments with a default!

````python
def compute_product(x, y=5):
    return x * y

# use the default value for y
print(compute_product(3))

# specify a y value other than the default
print(compute_product(5, 10))

# can also explicitly tell Python which arguments are which
print(compute_product(5, y=2))
print(compute_product(y=2, x=5))
````

---

### add documentation to a function

A good practice is to write documentation bout the function. 

````python
def compute_product(x, y=5):
    """
    This computes the product of x and y
    
    This is all part of the comment.
    """
    return x * y
````

#### Using the `numpy` module to manipulate array

&nbsp;&nbsp;&nbsp; NumPy is a powerful library for numerical computing in Python. It provides support for arrays, matrices, and a variety of mathematical functions to operate on these data structures.

---

##### create a numpy array

````python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print(arr)

# Create an array of zeros
zeros = np.zeros((3, 4))
print(zeros)

# Create an array of ones
ones = np.ones((2, 3))
print(ones)
````

---

````python
# Create an array with a range of values
range_array = np.arange(10)
print(range_array)

# Create an array with evenly spaced values
linspace_array = np.linspace(0, 1, 5)
print(linspace_array)
````

---

##### numpy array arithmetric operations

````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Addition
sum_array = a + b
print(sum_array)

# Subtraction
diff_array = a - b
print(diff_array)

# Multiplication
prod_array = a * b
print(prod_array)

# Division
div_array = a / b
print(div_array)
````

---

##### numpy array indexing and slicing

````python
#You can access and modify elements in a NumPy array using indexing and slicing:
arr = np.array([10, 20, 30, 40, 50])

# Accessing elements
print(arr[0])  # First element
print(arr[1:4])  # Slicing from index 1 to 3
````

##### Reshaping Arrays

````python

# You can change the shape of an array without changing its data:

arr = np.arange(12)
reshaped_arr = arr.reshape((3, 4))
print(reshaped_arr)
````

---

##### Mathematical Functions

````python
# NumPy provides a variety of mathematical functions:

arr = np.array([1, 2, 3, 4, 5])

# Mean
mean_value = np.mean(arr)
print("Mean:", mean_value)

# Standard Deviation
std_dev = np.std(arr)
print("Standard Deviation:", std_dev)

# Sum
total = np.sum(arr)
print("Sum:", total)
````

---

##### Matrix operations

````python
# NumPy also supports matrix operations:

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Matrix multiplication
C = np.dot(A, B)
print(C)

# Transpose of a matrix
transpose_A = A.T
print(transpose_A)
````

### Read and Write Text File

In Python, you can easily read from and write to text files using built-in functions. Below is a simple tutorial on how to do this.

---

#### Writing to a Text File

To write to a text file, you can use the `open()` function with the mode set to `'w'` (write mode). If the file does not exist, it will be created. If it does exist, it will be overwritten.

````python
# Define the file path
# Get the current working directory
import os

current_directory = os.getcwd()
file_path = os.path.join(current_directory, 'data.txt')

# Write to the file
with open(file_path, 'w') as file:
    file.write("Hello, World!\n") #`\n` means new line
    file.write("This is a text file.\n")
    file.write("Python makes it easy to read and write files.")
````

---


#### Reading from a Text File

To read from a text file, you can use the `open()` function with the mode set to `'r'` (read mode).

````python
# Read from the file
file_path = 'data.txt'
with open(file_path, 'r') as file:
    content = file.read()
    print(content)
````

### Read and write image file

In this tutorial, we will learn how to read and write PNG images using the `PIL` (Pillow) library and convert them to and from NumPy arrays. You may need to install the module, `pip install pillow`.


---

#### Read image
You can use Python to read an image file of local image file or a URL image file. The image reading is through the module of `PIL`, we will working on reading `geotiff` in following labs. 

````python
from PIL import Image
import numpy as np

# Load the image
url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg"
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# Convert the image to a NumPy array
img_array = np.array(img)

# Display the shape of the array
print("Image shape:", img_array.shape)
````

---

#### Writing a PNG Image
::: {.smaller}
To write a NumPy array back to a PNG image, we first convert the NumPy array back to a PIL image using `Image.fromarray()`, and then save it using the `save()` method.
:::

````python
# Load the image
url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg"
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# Convert the image to a NumPy array, we add a random value to each pixel
img_array = np.array(img) + np.random.rand(*img_array.shape)
# Set all values > 100 to 0
img_array[img_array > 100] = 0
# Convert to uint8
img_array_uint8 = img_array.astype(np.uint8)
# Convert the NumPy array back to a PIL image
img_from_array = Image.fromarray(img_array_uint8)

# # Save it
img_from_array.save('output_image.png')
````
